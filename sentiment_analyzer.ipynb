```python
import re
import string
from sklearn.metrics import classification_report, confusion_matrix 
from sklearn.model_selection import GridSearchCV 
```

```python
import pandas as pd
import numpy as np

tweet_df = pd.read_csv('tweet.csv',encoding="latin1", names=["target", "ids", "date", "flag", "user", "text"])
tweet_df.head(5)
```

```python
val = tweet_df['text'].values
target = tweet_df['target'].values
```

```python
print(val[:10])
print(type(val))
print(len(val))
print(type(target))
```

```python
lst = list(zip(val,target))
lst[:10]
df = pd.DataFrame(lst, columns = ['tweets','target']) 
print(df.shape)
df.head(5)
```

```python
#Cleaning the tweet column

#removal of url's

def url(text):
    text = re.sub(r"http\S+", "", text)
    return(text)
df['w/o_URL'] = df['tweets'].apply(lambda x: url(x))

#removal of punctuation
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    #text = re.sub('[0-9]+', '', text)
    return text

df['tweet'] = df['w/o_URL'].apply(lambda x: remove_punct(x))
df.head(10)
```

```python
#converts to lower case
def to_lower(text):
    return(text.lower())
df['tweets_lower'] = df['tweet'].apply(lambda x: to_lower(x))
df.head(10)
```

```python
def token(text):
    re.split('\W+', text)
    return text
df['token'] = df['tweets_lower'].apply(lambda x: token(x))
df.head(10)
```

```python
final_tweets = df[['target','tweets_lower']]
final_tweets.head(5)
```

```python
x_values = final_tweets['tweets_lower'].values
y_values = final_tweets['target'].values
```

```python
print(len(x_values))
print(len(y_values))
print(type(x_values))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_values, y_values, test_size=0.4, random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vector = CountVectorizer(encoding='latin-1',binary=False,min_df = 5 ,stop_words='english')
bi_vector = CountVectorizer(encoding='latin-1',stop_words='english',binary=False, ngram_range=(1,2),min_df = 5)
```

```python
x_train_vec = vector.fit_transform(x_train)
print(x_train_vec.shape)

x_test_vec = vector.transform(x_test)
print(x_test_vec.shape)
```

```python
from sklearn.svm import LinearSVC as svm
svm = svm(C=1, max_iter=1000)
svm.fit(x_train_vec,y_train)
```

```python
svm.score(x_test_vec, y_test)
```

```python
predictions = svm.predict(x_test_vec) 
print(classification_report(y_test, predictions))
```

```python
param_grid = {'C': [0.1, 1, 10, 100], 'max_iter' : [500, 1000, 1500]} 
grid = GridSearchCV(svm, param_grid, refit = True) 
  
# fitting the model for grid search 
grid.fit(x_train_vec, y_train) 
```
